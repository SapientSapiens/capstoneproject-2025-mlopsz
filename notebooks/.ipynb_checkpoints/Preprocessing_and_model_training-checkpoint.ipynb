### Let us load the feature engineered data and pre-process it for model training and testing

```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error, mean_squared_log_error
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import cross_val_score,cross_val_predict
from sklearn.ensemble import RandomForestRegressor
import lightgbm as lgb
```

```python
# loading the dataframe from the pickle file saved at the end of EDA
df = pd.read_pickle('./data/feature_engineered_data.pkl')
```

```python
df.head()
```

```python
df.info()
```

#### Let us indentify the continuous and categorical features

```python
categoricalFeatureNames = ["season","weather", "bad_weather", "holiday","workingday", "hour", "hour_workingday", "day_name","month","year", "rush_hour", "part_of_day", "temp_tolerance_feel"]
numericalFeatureNames = ["humidity","temp", "windspeed", "temp_hum_factor"]
```

```python
# setting categorical features datatype as str so that they are recognized by DictVectorizer for encoding
for feature in categoricalFeatureNames:
    df[feature] = df[feature].astype(str)
```

#### Let us first split this  feature engineered dataset into train, validation and test set

```python
df = df.set_index('datetime')
```

```python
# sanity-check
df.info()
```

```python
df_train = df[:'2012-04-30'] # 16 months out of 20
#df_train = df[:'2011-12-31'] # 12 months out of 20
#df_val = df['2012-01-01':'2012-04-30'] # 4 months out of 20
df_test = df['2012-05-01': '2012-08-31'] # 4 months out of 20

#df_full_train_new = df[:'2012-06-30'] # 18 months out of 20
#df_test_new = df['2012-07-01': '2012-08-31'] # 2 months out of 20
```

```python
X_train = df_train[categoricalFeatureNames + numericalFeatureNames]
y_train = df_train['count'].values
```

```python
X_test = df_test[categoricalFeatureNames + numericalFeatureNames]
y_test = df_test['count'].values
```

```python
# sanity-check
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

#### Encoding

```python
train_dict = X_train.to_dict(orient='records')
test_dict = X_test.to_dict(orient='records')
```

```python
# Initialize and fit DictVectorizer
dv = DictVectorizer(sparse=False)

X_train_vec = dv.fit_transform(train_dict)
X_test_vec = dv.transform(test_dict)
```

```python
dv.get_feature_names_out()
```

## Now let us train a few models with different algorithms to see which is best. Here we must remember how the target feature was tending towards normalized distribution when applied logarthmic transformations to it during the EDA. We train the models and evaluate the metrics

```python
# function for metrics evauation
def evaluateMetrics(y_true, y_pred):
  # Mean Squared Error
  mse_lr = mean_squared_error(y_true, y_pred)

  # Root Mean Squared Error
  rmse_lr = np.sqrt(mse_lr)

  # Mean Absolute Error
  mae_lr = mean_absolute_error(y_true, y_pred)

  # R-squared
  r2_lr = r2_score(y_true, y_pred)

  # Print Metrics
  print(f'MSE: {mse_lr:.4f}')
  print(f'RMSE: {rmse_lr:.4f}')
  print(f'MAE: {mae_lr:.4f}')
  print(f'R²: {r2_lr:.4f}')
```

#### Linear Regression

```python
# Iniatialize the RandomForrestRegressor
lr =  LinearRegression()

# apply logarithmic transformation to the target featrure
yTrainLog = np.log1p(y_train)

# Train the model
lr.fit(X_train_vec, yTrainLog)

# Make predictions
lpreds = lr.predict(X_test_vec)

# converting target feature back to original values for evaluating metrics
print ("RMSLE Value For Linear Regression: ", mean_squared_log_error(y_test,np.exp(lpreds)))
evaluateMetrics(y_test,np.exp(lpreds))
```

Random Forrest Regression

```python
# Find the Best n_estimators

# Store results
n_estimators_list = list(range(5, 101, 5))
r2_scores = []
rmse_scores = []

for n in n_estimators_list:
    rf_model = RandomForestRegressor(n_estimators=n, random_state=42, n_jobs=-1)

    # Train the model
    rf_model.fit(X_train_vec, np.log1p(y_train))

    # Predict using the validation dataset
    y_pred_rf = rf_model.predict(X_test_vec)
    y_pred_rf_nolog = np.expm1(y_pred_rf)

    # Compute R² score
    r2 = r2_score(y_test, y_pred_rf_nolog)
    r2_scores.append(r2)

    # Compute RMSE
    rmse = np.sqrt(mean_squared_error(y_test, y_pred_rf_nolog))
    rmse_scores.append(rmse)

    print(f'n_estimators = {n} -------> R² = {r2:.4f}, RMSE = {rmse:.4f}')
    print()
```

```python
# We start with n_estimators = 15 as found best in last step and evaluate the best max_depth value
# Find the Best max_depth
# Store results
n_estimators = 15
r2_scores = []
rmse_scores = []
depth_list = [5, 7, 9, 10, 11, 13, 15, 20, 25, 30, 35, 40, 45, 50, 60, 75, 90, 110, 130, 160]
for depth in depth_list:
    rf_model = RandomForestRegressor(n_estimators=n_estimators, random_state=42, max_depth = depth, n_jobs=-1)

    # Train the model
    rf_model.fit(X_train_vec, np.log1p(y_train))

    # Predict using the validation dataset
    y_pred_rf = rf_model.predict(X_test_vec)
    y_pred_rf_nolog = np.expm1(y_pred_rf)


    # Compute R² score
    r2 = r2_score(y_test, y_pred_rf_nolog)
    r2_scores.append(r2)

    # Compute RMSE
    rmse = np.sqrt(mean_squared_error(y_test, y_pred_rf_nolog))
    rmse_scores.append(rmse)

    print(f'depth = {depth} -------> R² = {r2:.4f}, RMSE = {rmse:.4f}')
    print()
```

```python
# We take the best evaluated max_depth = 40 and find the best min_samples_leaf value
# Find the Best min_samples_leaf
# Store results
n_estimators = 15
max_depth = 40
r2_scores = []
rmse_scores = []
leaf_samples_list =  [1, 3, 5, 6, 10, 15, 20, 50, 100, 200, 500]
for leaf_samples in leaf_samples_list:
    rf_model = RandomForestRegressor(n_estimators=n_estimators, random_state=42, max_depth = max_depth, min_samples_leaf = leaf_samples, n_jobs=-1)

    # Train the model
    rf_model.fit(X_train_vec, np.log1p(y_train))

    # Predict using the validation dataset
    y_pred_rf = rf_model.predict(X_test_vec)
    y_pred_rf_nolog = np.expm1(y_pred_rf)

    # Compute R² score
    r2 = r2_score(y_test, y_pred_rf_nolog)
    r2_scores.append(r2)

    # Compute RMSE
    rmse = np.sqrt(mean_squared_error(y_test, y_pred_rf_nolog))
    rmse_scores.append(rmse)

    print(f'leaf_samples = {leaf_samples} -------> R² = {r2:.4f}, RMSE = {rmse:.4f}')
    print()
```

We can see the best point is min_samples_leaf = 1 where R² is highest and RMSE is lowest. So best parameter value of the Random Forrest n_estimators = 15, max_depth = 40 and min_samples_leaf = 1

```python
# Train the hypertuned model
rf_tuned = RandomForestRegressor(
    n_estimators = 15,
    max_depth = 40,
    min_samples_leaf = 1,
    random_state=42,
    n_jobs=-1
)

rf_tuned.fit(X_train_vec, np.log1p(y_train))  # Use the actual rental counts (not log-transformed)

# Predict
y_pred_val_rf = rf_tuned.predict(X_test_vec)
y_pred_rf_nolog = np.expm1(y_pred_val_rf)

# evaluating and printing metrics for the Random Forrest model
evaluateMetrics(y_test, y_pred_rf_nolog)
print ("RMSLE Value For Random Forest Regression: ", mean_squared_log_error(y_test, y_pred_rf_nolog))
```

```python
# check feature importance of the model
feature_names = dv.get_feature_names_out()

# Grab importances from the trained RF
importances = rf_tuned.feature_importances_

# Pair them up and sort
feature_importance_dict = dict(zip(feature_names, importances))
sorted_importances = sorted(
    feature_importance_dict.items(),
    key=lambda x: x[1],
    reverse=True
)

# Print the top 20
for name, score in sorted_importances[:20]:
    print(f"{name:<40} {score:.4f}")
```

### LGBMRegressor

```python
#for n in range(100, 1001, 50): # ------->> best n_estimator find
#for n in range(1, 31, 1):  #------------>> best max_depth find
lgbm_model = lgb.LGBMRegressor(
    n_estimators=950,
    learning_rate=0.08,
    max_depth=10,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='poisson', # for avoiding negative predicted values
    random_state=42,
    verbose=-1
)

lgbm_model.fit(X_train_vec, np.log1p(y_train))
test_preds = lgbm_model.predict(X_test_vec)
#print(f'n_estimators = {n}')
evaluateMetrics(y_test, np.exp(test_preds))
print('RMSLE Value For LightGBM :', mean_squared_log_error(y_test,  np.exp(test_preds)))
```

```python
actual_pred_test_df = pd.DataFrame({
    'actual': y_test,
    'prediction': np.exp(test_preds)
})
actual_pred_test_df
```
